```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(data=df, x='Sentiment', hue='Sentiment', palette='coolwarm', legend=False)
plt.title("Public Sentimeter on Distance_Based Fare System")
plt.xlabel("Sentiment")
plt.ylabel("Number of Observation")
plt.show()
```

```python
plt.figure(figsize=(8,5))
emotion_count = df['Emotion'].value_counts().reset_index()
emotion_count.columns = ['Emotion', 'Count']
sns.barplot(data=emotion_count, x='Emotion', y='Count')  # Removed palette='muted'
plt.title('Emotion Frequencies')
plt.ylabel('Count')
plt.xlabel('Emotion')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
concern_location = df.groupby(['Location', 'Key Concern']).size().reset_index(name='counts')
top_concerns = df['Key Concern'].value_counts().nlargest(5).index
filtered = concern_location[concern_location['Key Concern'].isin(top_concerns)]
sns.barplot(data=filtered, x='Location', y='counts', hue='Key Concern')
plt.title('Top Key Concerns by Location')
plt.ylabel('Number of Comments')
plt.xlabel('Location')
plt.legend(title='Key Concern')
plt.show()
```

```python
commuter_sentiment = df.groupby(['Commuter Type', 'Sentiment']).size().unstack(fill_value=0)
commuter_sentiment_percent = commuter_sentiment.div(commuter_sentiment.sum(axis=1), axis=0)

commuter_sentiment_percent.plot(kind='bar', stacked=True, figsize=(10,6), colormap='tab20')
plt.title('Sentiment Distribution by Commuter Type')
plt.ylabel('Proportion of Comments')
plt.xlabel('Commuter Type')
plt.legend(title='Sentiment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
